### 1 - Visualisation d’un nuage de points et exploration des variables


```python
# 1 - pip install pandas numpy seaborn
import pandas as pd
import numpy as np


data = pd.read_csv('house_pricing.csv')
```

```python
# 2 - 
print("Ci dessous les varibles ainsi que leurs types : \n{}".format(data.dtypes))
```

```python
# 3 -
import matplotlib.pyplot as plt
data.plot.scatter(y='LotArea', x='SalePrice', title= "Gaphique : LotArea par rapport au SalesPrices", c = 'red')
```

```python
# 4 -
value = data.loc[(data['LotArea'] < 20000) & (data['SalePrice']<500000)]
value.plot.scatter(y='LotArea', x='SalePrice', title= "Gaphique des maisons avec un LotArea < 20 000 pieds carrés et un prix < 500 000$ ", c = 'green')
```

```python
# 5 -
data.plot.scatter(y='LotArea', x='LotFrontage', title= "Gaphique : LotArea par rapport au LotFrontage", c = 'blue')
```

```python
# 6 -
value = data.loc[(data['LotArea'] < 100000) & (data['LotFrontage']<200)]
value.plot.scatter(y='LotArea', x='LotFrontage', title= "Gaphique des maisons avec un LotArea < 100000 pieds carrés et un lotFontage < 200 pieds carrés ", c = 'green')
```

## 2 - Visualisation de relations continues 

```python
# 1 et 2 -
import seaborn as sns
df = pd.read_csv('sales_predictions.csv')
data2 = sns.relplot(data=df, y='item_price', x='date').set(title='Prix en fonction du temps')
```

Le prix reste relativemennt constant au fur et à mesure du temps

```python
# 3 -
from random import sample
data2 = sns.relplot(data=df.sample(n=50, random_state=1), y='item_price', x='date').set(title='Prix en fonction du temps')
```

```python
# 4 -
data2 = sns.relplot(data=df.sample(n=50, random_state=1), y='item_price', x='date', sizes=(20, 6)).set(title='Prix en fonction du temps')
```
